In this assignment students have to transform iris data into 3 dimensions and plot a 3D chart with transformed dimensions and colour each data point with specific class.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(data = np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['Species'])
df.head()
```

```python
t = iris.target
color = np.choose(t, ['r', 'b', 'black'])
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(7,6))
ax = plt.axes(projection='3d')

x = df['sepal length (cm)']
y = df['sepal width (cm)']
z = df['petal length (cm)']

ax.scatter(x,y,z,c=color)

ax.set_xlabel('Sepal lenght')
ax.set_ylabel('Sepal width')
ax.set_zlabel('Petal lenght')

setosa = mpatches.Patch(color='r', label='Setosa')
versicolor = mpatches.Patch(color='b', label='Versicolor')
virginica = mpatches.Patch(color='black', label='Virginica')

plt.legend(handles=[setosa,versicolor,virginica])
```
